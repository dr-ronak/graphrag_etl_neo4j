```python
prompt = """
Provide an ontology for RDF knowledge graph which has data related to 'personal loan' in 'cards and payments' domain.
Please include 'customer segmentation','risk for loan default','financial regulations','credit score','credit exposure'. 
Give all ontologies with their respective taxonomy. 
"""

response = openai.ChatCompletion.create(
  engine='CapOneGpt4',
  messages = [{"role":"system","content":"You are a business domain expert in cards and payments."},
              {"role":"user","content":prompt}],
  temperature=0.5,
  max_tokens=30000,
  top_p=0.6,
  frequency_penalty=0,
  presence_penalty=0,
  stop=None)
```

```python
#RDF with specification of 'customer segmentation','risk for loan default','financial regulations','credit score','credit exposure'
print(response.choices[0].message.content)
```
